Using dictionary

```python
climate_data = {
    'Date': ['1/1/2023', '1/1/2024','1/1/2025'],
    'temperature': [27,28,29],
    'windspeed':[3,2,5],
    'event': ['cloudy','rainny','sunny']
}
df = pd.DataFrame(climate_data)
df
```

Using Tuple List

```python
weather_data = [
    ('1/1/2023',27,3,'cloudy'),
    ('1/1/2024',28,3,'rainny'),
    ('1/1/2025',29,3,'sunny'),
]
df = pd.DataFrame(data= weather_data, columns=['date','temperature','windspeed','event'],index=['x','y','z'],)
df
```

Using List of dictionaries

```python
climate_data = [
    {'date':'1/1/2023', 'temperature':27, 'winnspeed':3, 'event':'Cloudy'},
    {'date':'1/1/2024', 'temperature':28, 'winnspeed':7, 'event':'rainny'},
    {'date':'1/1/2025', 'temperature':29, 'winnspeed':9, 'event':'sunny'},
]
df = pd.DataFrame(data=climate_data, columns=['date','temperature','windspeed','event'])
df
```
