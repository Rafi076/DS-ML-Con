```python
# Convert 'Percentage' column to numeric after removing '%' symbol
df['Percentage'] = df['Percentage'].str.rstrip('%').astype(float)

# Assign grades based on conditions
conditions = [
    df['Percentage'] > 70,
    (df['Percentage'] <= 70) & (df['Percentage'] >= 60),
    (df['Percentage'] < 60) & (df['Percentage'] >= 50),
    df['Percentage'] < 50
]

grades = ['A+', 'B+', 'C+', 'FAIL']

# Apply the conditions and assign to the 'Grade' column
df['Grade'] = np.select(conditions, grades)

df

# # or we can do this way -- 
# df.loc[df['Percentage']=='70%',['Grade']] = 'A+'
# df.loc[df['Percentage']<'50%',['Grade']] = 'FAIl'
```
