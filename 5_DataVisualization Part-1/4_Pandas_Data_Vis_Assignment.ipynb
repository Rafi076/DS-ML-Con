```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
```

```python
# Code goes here!
entries = df.shape[0]  
print(f"Number of entries: {entries}")
```

```python
# Code goes here!
df.head()
```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!
plt.figure(figsize=(10, 6))
plt.scatter(df['c'], df['d'], c='green', s=50, alpha=0.8)
plt.show()
```

** Create a histogram of the 'd' column.**

```python
# Code goes here!
plt.figure(figsize=(8, 5))
plt.hist(df['d'], bins=10, color='blue',alpha=0.8)
plt.show()
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!
plt.style.use('bmh')
plt.figure(figsize=(8, 5))
plt.hist(df['d'], bins=50, color='purple', edgecolor='grey', alpha=0.7)
plt.show()
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!
plt.figure(figsize=(10, 6))
plt.scatter(df['a'], df['b'], c='green', s=df['d'] * 200, alpha=0.9)
plt.show()
```

**Recreate the plot below**

```python
# Code goes here!
# Set the figure size
plt.figure(figsize=(12, 8))

# Create a hexbin plot of 'a' vs 'b', colored by 'd'
plt.hexbin(df['a'], df['b'], C=df['d'], gridsize=30, cmap='coolwarm', alpha=0.7, edgecolors='none')

# Add a colorbar to show the color scale
plt.colorbar(label='Counts')

# Display the plot
plt.show()

# plt.hexbin(x, y, gridsize=30, cmap='coolwarm', edgecolors='none')
# plt.colorbar(label='Counts')
```

**Create a boxplot comparing all columns in the data.**<br>
Hint: use `color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')` and `sym = 'r+'` 

```python
# Code goes here!
# Set the figure size
plt.figure(figsize=(10, 7))

# Custom colors for the boxplot elements
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')

# Create the boxplot
df.boxplot(color=color, sym='r+')

# Adding title and labels
plt.title('Boxplot of All Columns', fontsize=16)
plt.ylabel('Values', fontsize=14)

# Display the plot
plt.show()
```
