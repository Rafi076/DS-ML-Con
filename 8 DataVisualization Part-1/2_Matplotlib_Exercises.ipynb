```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, f_oneway

# Mock dataset creation based on report analysis
data = {
    "Year": ["First-Year"] * 22 + ["Senior-Year"] * 42,
    "Academic Challenges": [75, 80, 78, 82, 85, 77, 79, 84, 80, 78, 76, 83, 81, 79, 82, 80, 81, 76, 85, 80, 77, 83] +
                           [68, 70, 72, 65, 67, 66, 71, 69, 74, 68, 73, 67, 72, 66, 70, 68, 74, 67, 72, 65, 68, 66, 71, 69, 74, 70, 68, 65, 67, 73, 66, 72, 65, 68, 70, 69, 74, 68, 72, 65, 68, 66],
    "Faculty Interactions": [82, 85, 79, 84, 83, 81, 80, 85, 78, 76, 83, 82, 81, 79, 84, 82, 83, 76, 86, 81, 77, 84] +
                            [70, 72, 74, 68, 69, 71, 73, 75, 72, 74, 70, 71, 73, 68, 69, 71, 75, 70, 74, 68, 69, 72, 74, 75, 71, 70, 68, 72, 71, 75, 74, 70, 68, 71, 72, 74, 73, 75, 70, 69, 68, 71],
    "Satisfaction": [85, 88, 86, 87, 90, 84, 85, 89, 87, 84, 86, 88, 89, 86, 90, 87, 88, 83, 90, 87, 84, 88] +
                     [75, 78, 76, 74, 75, 77, 78, 76, 75, 74, 77, 76, 78, 75, 74, 76, 77, 75, 76, 74, 75, 77, 78, 76, 75, 74, 75, 77, 78, 76, 75, 74, 76, 75, 77, 78, 75, 76, 74, 75, 77, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Analysis 1: ANOVA for engagement levels between groups
anova_result = f_oneway(
    df[df["Year"] == "First-Year"]["Academic Challenges"],
    df[df["Year"] == "Senior-Year"]["Academic Challenges"]
)

# Analysis 2: Correlation between engagement and satisfaction
correlation_academic, p_value_academic = spearmanr(df["Academic Challenges"], df["Satisfaction"])
correlation_faculty, p_value_faculty = spearmanr(df["Faculty Interactions"], df["Satisfaction"])

# Visualization 1: Boxplot for engagement levels by year
plt.figure(figsize=(8, 6))
sns.boxplot(x="Year", y="Academic Challenges", data=df, palette="muted")
plt.title("Engagement Levels: Academic Challenges by Year")
plt.xlabel("Year")
plt.ylabel("Academic Challenges")
plt.tight_layout()
plt.show()

# Visualization 2: Results Bar Plot
results_data = {
    "Metric": ["ANOVA F-statistic", "ANOVA p-value", "Spearman Correlation (Academic Challenges)", "Spearman Correlation (Faculty Interactions)"],
    "Value": [anova_result.statistic, anova_result.pvalue, correlation_academic, correlation_faculty]
}
results_df = pd.DataFrame(results_data)

plt.figure(figsize=(10, 6))
sns.barplot(y="Metric", x="Value", data=results_df, palette="coolwarm")
plt.title("Statistical Results")
plt.xlabel("Value")
plt.ylabel("Metric")
plt.tight_layout()
plt.show()

# Print Results
print("ANOVA Results: F-statistic =", anova_result.statistic, ", p-value =", anova_result.pvalue)
print("Spearman Correlation (Academic Challenges vs Satisfaction):", correlation_academic, ", p-value =", p_value_academic)
print("Spearman Correlation (Faculty Interactions vs Satisfaction):", correlation_faculty, ", p-value =", p_value_faculty)
```
